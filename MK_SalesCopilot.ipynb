# Creación de la Base de Conocimiento

En este módulo se construye la base de conocimiento que utilizará el agente.

Proceso:

1. Cargar los documentos.
2. Dividirlos en fragmentos (chunks).
3. Crear los embeddings.
4. Almacenar los vectores en ChromaDB.
5. Verificar el funcionamiento del recuperador.
6. Carga base de clientes

```python
# ============================================================
# RUTAS DE LOS DOCUMENTOS
# ============================================================

RUTA_DOCUMENTOS = [
    "data/catalogo.pdf",
    "data/experta_producto.pdf"
]
```

```python
!pip install pypdf
```

```python
# ============================================================
# CARGA DE DOCUMENTOS
# ============================================================

import os
from langchain_community.document_loaders import PyPDFLoader

documentos = []

for archivo in RUTA_DOCUMENTOS:
    if os.path.exists(archivo):
        try:
            loader = PyPDFLoader(archivo)
            paginas = loader.load()
            documentos.extend(paginas)
            print(f"✅ Cargado: {os.path.basename(archivo)} ({len(paginas)} páginas)")
        except Exception as e:
            print(f"❌ Error al cargar {os.path.basename(archivo)}: {e}")
    else:
        print(f"⚠️ Archivo no encontrado: {archivo}")

print(f"\nTotal general: Se cargaron {len(documentos)} páginas.")
```

```python
!pip install openpyxl
```

```python
# ============================================================
# CARGAR BASE DE CLIENTES
# ============================================================
df_clientes = pd.read_excel("data/clientes.xlsx", engine="openpyxl")

print(f"Clientes cargados: {len(df_clientes)}")
df_clientes.head()
```

```python
# ============================================================
# CONVERTIR CLIENTES A DOCUMENTOS
# ============================================================

clientes_docs = []

for _, fila in df_clientes.iterrows():

    texto = f"""
Cliente: {fila['NOMBRE']}
Celular: {fila['CELULAR']}
Ciudad: {fila['CIUDAD']}
Tipo de piel: {fila['TIPO PIEL']}
Tono de base: {fila['TONO BASE']}
"""

    clientes_docs.append(
        Document(
            page_content=texto,
            metadata={
                "cliente": str(fila["NOMBRE"]),
                "ciudad": str(fila["CIUDAD"])
            }
        )
    )

print(f"✅ Clientes convertidos: {len(clientes_docs)}")
```

```python
# ============================================================
# CHUNKING
# ============================================================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

todos_los_documentos = documentos + clientes_docs

chunks = text_splitter.split_documents(todos_los_documentos)

print(f"Se generaron {len(chunks)} fragmentos.")
```

```python
# ============================================================
# CREAR BASE VECTORIAL
# ============================================================

vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="chroma_db"
)

print(" Base vectorial creada correctamente.")
```

```python
retriever = vector_db.as_retriever(
    search_kwargs={"k": 4}
)

print("Retriever creado correctamente.")
```

# System Prompt

Se define el comportamiento del agente experto.

El agente debe responder únicamente utilizando la información recuperada desde la base de conocimiento.

Si la información no está disponible en los documentos, debe indicarlo claramente sin inventar respuestas.

```python
from langchain_core.prompts import ChatPromptTemplate

# Definición limpia y directa de la plantilla
prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        """
Eres una asesora experta de productos Mary Kay.

Responde únicamente utilizando la información recuperada de la base de conocimiento.

Si la información no está disponible responde:

'No encontré información suficiente en la base de conocimiento.'

Sé clara, profesional y amable.
"""
    ),
    (
        "human",
        """
Contexto:

{context}

Pregunta:

{question}
"""
    )
])

print("✅ Plantilla de Mary Kay configurada correctamente.")
```

```python
# ============================================================
# PLANTILLA DEL PROMPT
# ============================================================

# No duplicamos nada, el objeto 'prompt' de la celda anterior ya está listo.
print("Prompt creado correctamente.")
print(f"Variables esperadas por el prompt: {prompt.input_variables}")
```

# Función RAG

Se implementa la lógica de Recuperación Aumentada por Generación (RAG).

El proceso consiste en:

1. Recuperar los documentos más relevantes desde ChromaDB.
2. Construir el contexto.
3. Enviar el contexto junto con la pregunta al modelo Gemini.
4. Obtener una respuesta basada únicamente en la información recuperada.

```python
# ============================================================
# FUNCIÓN RAG
# ============================================================

def responder(pregunta):

    documentos = retriever.invoke(pregunta)

    contexto = "\n\n".join(
        doc.page_content
        for doc in documentos
    )

    mensajes = prompt.invoke({
        "context": contexto,
        "question": pregunta
    })

    respuesta = llm.invoke(mensajes)

    return {
        "respuesta": respuesta.content,
        "documentos": documentos
    }
```

```python
# Convertir el RAG en una herramienta

from langchain.tools import tool

@tool
def buscar_en_catalogo(pregunta: str) -> str:
    """
    Busca información en la base de conocimiento de Mary Kay.
    """

    resultado = responder(pregunta)

    return resultado["respuesta"]
```

# Agente con LangGraph

En este módulo se construye el flujo del agente utilizando LangGraph.

El agente recibe una pregunta del usuario, consulta la base de conocimiento mediante RAG y devuelve una respuesta utilizando Gemini.

```python
from typing import TypedDict

from langgraph.graph import StateGraph
from langgraph.graph import START, END


class AgentState(TypedDict):
    question: str
    answer: str


def agente(state: AgentState):

    pregunta = state["question"]

    resultado = responder(pregunta)

    return {
        "answer": resultado["respuesta"]
    }


builder = StateGraph(AgentState)

builder.add_node("agente", agente)

builder.add_edge(START, "agente")

builder.add_edge("agente", END)

graph = builder.compile()

print("Agente creado correctamente.")
```

```python
config = {
    "configurable": {
        "thread_id": "cliente_001"
    }
}
```

```python
respuesta = graph.invoke(
    {
        "question": "¿Qué beneficios tiene TimeWise?"
    },
    config=config
)

print(respuesta["answer"])
```

```python
respuesta = graph.invoke(
    {
        "question": "¿Y para qué tipo de piel sirve?"
    },
    config=config
)

print(respuesta["answer"])
```

# Memoria de conversación

En este módulo se incorpora memoria al agente para mantener el contexto entre preguntas consecutivas.

Esto permite realizar preguntas de seguimiento sin repetir toda la información.

```python
# ============================================================
# MEMORIA DE CONVERSACIÓN
# ============================================================

from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()

graph = builder.compile(
    checkpointer=memory
)

config = {
    "configurable": {
        "thread_id": "chat_marykay"
    }
}

print("Memoria creada correctamente.")
```

# Chat interactivo

Esta celda permite mantener una conversación con el agente utilizando la memoria implementada en el módulo anterior.

```python
print("="*60)
print("MK SALES ASSISTANT")
print("="*60)

config = {
    "configurable":{
        "thread_id":"chat_mk"
    }
}

while True:

    pregunta = input("Usuario: ")

    if pregunta.lower()=="salir":
        break

    respuesta = graph.invoke(
        {
            "question": pregunta
        },
        config=config
    )

    print("\nAsistente:")
    print(respuesta["answer"])
```
